```python
from DQN_Agent_offline import DQN_Agent
import pandas as pd; import numpy as np 

path  = 'RL_Data/07_Offline/03_Data/'

agent_state_size = 5
# Agent_1 = DQN_Agent(agent_state_size, 2)

# df_offline = pd.read_csv(path+'offline_rbc_df.csv',converters={'state': pd.eval}) 
df_offline = pd.read_csv(path+'offline_rbc_df.csv') 
```

```python
df_offline['state_m'] = df_offline['state'].map(lambda x: x.replace('\n','').replace('[','').replace(']','').split())
df_offline['next_state_m'] = df_offline['next_state'].map(lambda x: x.replace('\n','').replace('[','').replace(']','').split())

s0,s1,s2,dr_m1, dr_0 = [],[],[],[],[]
for i in range(len(df_offline)):
     s0.append(float(df_offline['state_m'].iloc[i][0]))
     s1.append(float(df_offline['state_m'].iloc[i][1]))
     s2.append(float(df_offline['state_m'].iloc[i][2]))
     dr_0.append(float(df_offline['state_m'].iloc[i][-1]))
     dr_m1.append(float(df_offline['state_m'].iloc[i][-2]))

df_offline["s0"]= s0
df_offline["s1"]= s1
df_offline["s2"]= s2
df_offline["s3"]= dr_m1
df_offline["s4"]= dr_0


s0,s1,s2,dr_m1, dr_0 = [],[],[],[],[]
for i in range(len(df_offline)):
     s0.append(float(df_offline['next_state_m'].iloc[i][0]))
     s1.append(float(df_offline['next_state_m'].iloc[i][1]))
     s2.append(float(df_offline['next_state_m'].iloc[i][2]))
     dr_0.append(float(df_offline['next_state_m'].iloc[i][-1]))
     dr_m1.append(float(df_offline['next_state_m'].iloc[i][-2]))

df_offline["n_s0"]= s0
df_offline["n_s1"]= s1
df_offline["n_s2"]= s2
df_offline["n_s3"]= dr_m1
df_offline["n_s4"]= dr_0
df_offline["done"]=False


df_offline['off_next_state'] = df_offline[["n_s0","n_s1","n_s2","n_s3","n_s4"]].values.tolist()
df_offline['off_state'] = df_offline[["s0","s1","s2","s3","s4"]].values.tolist()


```
